## Univariate Exploration

I'll start by looking at the distribution of the main variable of interest: borrower_apr

## Bivariate Exploration

> To start off with, I want to look at the pairwise correlations present between features in the data.

```python
numeric_vars = ['borrower_apr','borrower_rate','estimated_effective_yield', 'estimated_return','estimated_loss']
categoric_vars = ['term', 'prosper_score','prosper_rating_alpha']
```

```python
# plot matrix: sample 500 loans so that plots are clearer and they render faster
print("loan_data.shape=",loan_data.shape)
loan_data_samp = loan_data.sample(n=500, replace = False)
print("loan_data_samp.shape=",loan_data_samp.shape)

# correlation plot
plt.figure(figsize = [12, 10])
sb.heatmap(loan_data_samp[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0);
```

The Correlation between the Borrower APR and the other features such as Borrower Rate, Estimated Effective Yield, Estimated Return and Estimated Loss are all positively correlated.

This means that an increase in the Borrower APR will lead to an increase in all these features.

```python
g = sb.PairGrid(data = loan_data, vars = numeric_vars)
g = g.map_diag(plt.hist, bins=20)
g.map_offdiag(plt.scatter, alpha=0.2);
```

```python
loan_data_samp = loan_data.sample(n=2000, replace = False)


def boxgrid(x, y, **kwargs):
    """ Quick hack for creating box plots with seaborn's PairGrid. """
    default_color = sb.color_palette()[0]
    sb.boxplot(x=x, y=y, color=default_color)

plt.figure(figsize = [10, 10])
g = sb.PairGrid(data = loan_data_samp, y_vars = numeric_vars, 
                                        x_vars = categoric_vars,
                height = 3, aspect = 1.5)
g.map(boxgrid)
plt.show();
```

The figure shows that the borrower APR decreases with the better rating. Borrowers with the best Prosper ratings have the lowest APR. It means that the Prosper rating has a strong effect on borrower APR. Borrowers with better prosper rating also have lower estimated losses, estimated effective yields and lower borrower rate.

Finally, let's look at relationships between the categorical features

```python
plt.figure(figsize = [8, 10])

# subplot 1: Prosper rating vs term
plt.subplot(4, 1, 1)
sb.countplot(data = loan_data, x = 'prosper_rating_alpha', hue = 'term', palette = 'Blues')

# subplot 2: is_borrower_home_owner vs. term
ax = plt.subplot(4, 1, 2)
sb.countplot(data = loan_data, x = 'is_borrower_home_owner', hue = 'term', palette = 'Blues')
plt.xticks(rotation = 15)


# subplot 3: Prosper rating vs. employment status, use different color palette
ax = plt.subplot(4, 1, 3)
sb.countplot(data = loan_data, x = 'prosper_rating_alpha', hue = 'is_borrower_home_owner', palette = 'Greens')
ax.legend(loc = 1, ncol = 2); # re-arrange legend to remove overlapping
plt.xticks(rotation = 15);
```

Most borrowers chose to go for loans with term of 36, while less borrowers chose the 60 term loans while very few borrowers chose the 12months term loan

## Multivariate Exploration

> The main thing I want to explore in this section of the analysis is how the categorical variables (Loan Terms and Prosper rating) play into the relationship between borrower APR and Rate.

```python
# Term effect on relationship of Borrower APR and Rate
g=sb.FacetGrid(data=loan_data, aspect=1.2, height=5, col='term', col_wrap=4)
g.map(sb.regplot, 'borrower_apr', 'borrower_rate', x_jitter=0.04, scatter_kws={'alpha':0.1});
g.add_legend();
```

Term doesn't seem to have effect on relationship of Borrower APR and Rate

```python
# Prosper rating effect on relationship of Borrower APR and Estimated Return
g=sb.FacetGrid(data=loan_data, aspect=1.2, height=5, col='prosper_rating_alpha', col_wrap=4)
g.map(sb.regplot, 'borrower_apr', 'estimated_return', x_jitter=0.04, scatter_kws={'alpha':0.1});
g.add_legend();
```

The Estimated Return seems to be decreasing with better Prosper Rating. While the Borrower APR is slightly unchanged.

```python
fig = plt.figure(figsize = [8,6])
ax = sb.pointplot(data = loan_data, x = 'prosper_rating_alpha', y = 'borrower_apr', hue = 'term',
           palette = 'Blues', linestyles = '', dodge = 0.4, ci='sd')
plt.title('Borrower APR across Prosper rating and term')
plt.ylabel('Mean Borrower APR')
ax.set_yticklabels([],minor = True);
```

Interestingly, the borrower APR decrease with the increase of borrow term for people with HR-C raings. But for people with B-AA ratings, the APR increase with the increase of borrow term.

```python
fig, ax = plt.subplots(ncols=2, figsize=[12,6])
sb.pointplot(data = loan_data, x = 'prosper_rating_alpha', y = 'estimated_loss', hue = 'term',
           palette = 'Purples', linestyles = '', dodge = 0.4, ax=ax[0])
sb.pointplot(data = loan_data, x = 'prosper_rating_alpha', y = 'estimated_return', hue = 'term',
           palette = 'Greens', linestyles = '', dodge = 0.4, ax=ax[1]);
```
